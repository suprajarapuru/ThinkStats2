```python
# DSC 530
# Programming Assignment Week 3 1.2
# Author Supraja Rapuru
# 6/25/2021

from __future__ import print_function

import numpy as np
import sys

import nsfg
import thinkstats2


def ReadFemResp(dct_file,dat_file,nrows=None):
    dct = thinkstats2.ReadStataDct(dct_file)
    df = dct.ReadFixedWidth(dat_file, compression='gzip', nrows=nrows)
    return df


def ReadFemPreg(dct_file='2002FemPreg.dct',
                dat_file='2002FemPreg.dat.gz'):
    dct = thinkstats2.ReadStataDct(dct_file)
    df = dct.ReadFixedWidth(dat_file, compression='gzip')
    return df


def ValidatePregnum(resp):
    preg = nsfg.ReadFemPreg()
    preg_map = nsfg.MakePregMap(preg)
    for index, pregnum in resp.pregnum.items():
        caseid = resp.caseid[index]
        indices = preg_map[caseid]
        if len(indices) != pregnum:
            print(caseid, len(indices), pregnum)
            return False

    return True


def main():
    dct_file='2002FemResp.dct'
    dat_file='2002FemResp.dat.gz'
    # reading the respondent file
    print("Started Reading...Respondent File")
    resp = ReadFemResp(dct_file,dat_file)
    print("Completed Reading...Respondent File")
    
    # validate the respondent file by comparing total resp
    # This comparison is from known values
    assert(len(resp) == 7643)
    assert(resp.pregnum.value_counts()[1] == 1267) 
    
    # Displaying pregnum
    print(resp.pregnum.value_counts().sort_index())
       
    # Cross Validate 
    assert(ValidatePregnum(resp))

    print('All tests passed.')
    
   
if __name__ == '__main__':
    main()
```
